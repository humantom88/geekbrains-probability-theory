## Задание 1

Из колоды в 52 карты извлекаются случайным образом 4 карты. 

a) Найти вероятность того, что все карты – крести. 

б) Найти вероятность, что среди 4-х карт окажется ХОТЯ БЫ один туз.

### Решение

а) Найти вероятность того, что все карты – крести.

Вероятность вытащить первую карту масти крести:
$P_1 = \frac{13}{52} = \frac{1}{4}$

Вероятность вытащить вторую карту масти крести:
$P_2 = \frac{12}{51}$

Вероятность вытащить третью карту масти крести:
$P_3 = \frac{11}{50}$

Вероятность вытащить четвертую карту масти крести:
$P_4 = \frac{10}{49}$

```python
P = 1/4 * 12/51 * 11/50 * 10/49
P
```

б) Найти вероятность, что среди 4-х карт окажется ХОТЯ БЫ один туз.

Вероятность, что первая карта - не туз: $P_1 = 48/52$

Вероятность, что вторая карта - не туз: $P_2 = 48/51$

Вероятность, что третья карта - не туз: $P_2 = 48/50$

Вероятность, что четвертая карта - не туз: $P_2 = 48/49$

Вероятность, что хотя бы один туз $P_{Tуз} = 1 - P_{неТуз}$

```python
P = 1 - (48/52 * 48/51 * 48/50 * 48/49)
P
```

## Задание 3

В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

### Решение

Три последовательных события

Вероятность, что первая деталь окрашена: $P_1 = 9/15$

Вероятность, что вторая деталь окрашена: $P_1 = 8/14$

Вероятность, что третья деталь окрашена: $P_1 = 7/13$

Произведение вероятностей даст нам искомую вероятность

```python
P = 9/15 * 8/14 * 7/13
P
```

## Задание 4

В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

Вероятность, что первый билет будет выигрышным: $P_1 = 1/50$

Вероятность, что второй билет будет выигрышным: $P_2 = 1/99$

Итоговая вероятность:

```python
P = 1/50 * 1/99
P
```
